#### Q4. In python, plot the probability mass function and cumulative distribution function of the following random variables

(a) Binomial, n = 30, p = 0.2
(b) Geometric, k = 1 . . . , 20, p = 0.2

```python
from scipy import stats 
import matplotlib.pyplot as plt 
```

```python
X = stats.binom(30, 0.2) # Declare X to be a binomial random variable
arr = [x for x in range(1, 31)]
answer = []
for i in range(1, 31):
    answer.append(X.cdf(i))  # P(X <= i)

plt.plot([x for x in range(1, 31)], answer)
plt.suptitle('Cumulative Distribution')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
answer = []
for i in range(1, 31):
    answer.append(X.pmf(i))  # P(X <= i)

plt.plot([x for x in range(1, 31)], answer)
plt.suptitle('Mass Distribution')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

## Solving B part

```python
X = stats.geom(0.2) # Declare X to be a geometric random variable

answer = []

for i in range(1, 21):
    answer.append(X.cdf(i))

plt.plot([x for x in range(1, 21)], answer)
plt.suptitle('Cumulative Distribution')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
answer = []

for i in range(1, 21):
    answer.append(X.pmf(i))

plt.plot([x for x in range(1, 21)], answer)
plt.suptitle('Mass Distribution')
plt.xlabel('')
plt.ylabel('')
plt.show()
```
