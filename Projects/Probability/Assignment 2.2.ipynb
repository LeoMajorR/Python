#### Q2. Write a python code to implement monte carlo integration of

$\int\limits_0^1$ $\sqrt{sin(x)}$dx.

Note: please report a distribution and not a single value.

```python
%matplotlib inline
from scipy import random
import numpy as np 
import matplotlib.pyplot as plt 
import math
```

```python
a = 0
b = 1

N = 10000

xrand  = random.uniform(a, b, N)

def fun(x):
    l = np.sin(x)
    return np.sqrt(l)

integral = 0.0
for i in range(N):
    integral += fun(xrand[i])

answer = (b-a)/float(N)*integral
print("The integral is > ", answer)
```

```python
a = 0
b = 1
N = 1000

def func(x):
    l = np.sin(x)
    return np.sqrt(l)

areas = []

for i in range(N):
    xrand = np.zeros(N)

    for i in range(len(xrand)):
        xrand[i] = random.uniform(a, b)
        integral = 0.0
    
    for i in range(N):
        integral += func(xrand[i])


    answer = (b-a)/float(N)*integral

    areas.append(answer) 

plt.title("Distribution of Areas Calculator")
plt.hist(areas, bins = 30, ec = "black")
plt.xlabel("Areas")
plt.plot()
```
