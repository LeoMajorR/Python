## Conditional Probability


#### Q2. Repeat the above exercise with python and report the distribution obtained over several runs.


```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from time import time as TT
from random import choice as ch
import numpy as np
from scipy import stats 
```

```python
ac = []
tc = []
N = []
st = TT()
for M in range(1,1000): #Outer loop from 1 to 10000
    st1 = TT()
    score = []
    runs = 0
    cards = [1,2,3,4]
    for K in range(1,M): # sub loop that simulates 1 to M(outerloop) games
        aset = []
        host = cards.copy()
        hbk = ch(host) #Randomly choose as answer which host knows
        aset.append(hbk)
        #print("The host knows the answer",hbk)
        player = cards.copy()
        px = ch(player) # Contestanrs random guess
        aset.append(px)
        #print ("Players first choice",px)
        chance = 0
        for i in host: # The computation....host will eliminate P(X|DOOR) = 0
            if i not in aset:
                chance = i
        #print ("The elimination",chance)
        #print (player)
        player.pop(player.index(chance))
        player.pop(player.index(px))
        #print ("final answe",player)
        if player[0] == hbk:
            score.append(1)
        else:
            score.append(0)
        runs = K
        #print ("\n\n")
    ac.append(np.mean(score))
    N.append(M)
    en1 = TT()
    tc.append(en1-st1)
en = TT()
```

```python
plt.xlabel("Number of Games Played")
plt.ylabel("Number of Games Won")

plt.plot(N,ac)
plt.show()

plt.plot(N,tc)
plt.show()
```

#### Q4. In python, plot the probability mass function and cumulative distribution function of the following random variables

(a) Binomial, n = 30, p = 0.2


```python
X = stats.binom(30, 0.2) # Declare X to be a binomial random variable
arr = [x for x in range(1, 31)]
answer = []
for i in range(1, 31):
    answer.append(X.cdf(i))  # P(X <= i)

plt.plot([x for x in range(1, 31)], answer)
plt.suptitle('Cumulative Distribution')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.show()
```

```python
answer = []
for i in range(1, 31):
    answer.append(X.pmf(i))  # P(X <= i)
    
val = [x for x in range(1, 31)]
plt.bar(val, answer)
plt.suptitle('Mass Distribution')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.show()
```
