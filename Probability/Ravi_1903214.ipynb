## Conditional Probability


#### Q2. Repeat the above exercise with python and report the distribution obtained over several runs.


```python
%matplotlib inline
import matplotlib.pyplot as plt
from random import choice as ch
import numpy as np
from scipy import stats 
import random
```

```python
# Make 1000 trials.
n_tries = 1000
# Lists to store results from stay and switch strategy


def MonteHall():
    stay_results = []
    switch_results = []
    for i in range(n_tries):
        # Same code as above, for one trial
        doors = ['car', 'goat', 'goat','goat']
        random.shuffle(doors)
        my_door_index = random.choice([0, 1, 2, 3])
        stay_result = doors.pop(my_door_index)
        doors.remove('goat')
        doors.remove('goat')
        switch_result = doors[0]
        # Put results into result lists
        stay_results.append(stay_result)
        switch_results.append(switch_result)
    return (stay_results.count('car') / n_tries, switch_results.count('car') / n_tries)
a=[]
b=[]
for i in range(1000):
    x,y = MonteHall()

    a.append(x)
    b.append(y)
plt.title("Distribution Without Switching")
plt.plot(a, color="red")
plt.xlabel("No of Games Played")
plt.ylabel("Frequency of Game Won")
plt.plot()
```

```python
plt.title("Distribution With Switching")
plt.plot(b, color="green")
plt.xlabel("No. of Games Played")
plt.ylabel("Frequency of Game Won")
plt.plot()
```

#### Q2. Write a python code to implement monte carlo integration of

$\int\limits_0^1$ $\sqrt{sin(x)}$dx.

Note: please report a distribution and not a single value.

```python
a = 0
b = 1

N = 10000

xrand  = random.uniform(a, b, N)

def fun(x):
    l = np.sin(x)
    return np.sqrt(l)

integral = 0.0
for i in range(N):
    integral += fun(xrand[i])

answer = (b-a)/float(N)*integral
print("The value integral is > ", answer, "Unit")
```

```python
a = 0
b = 1
N = 1000

def func(x):
    l = np.sin(x)
    return np.sqrt(l)

areas = []

for i in range(N):
    xrand = np.zeros(N)

    for i in range(len(xrand)):
        xrand[i] = random.uniform(a, b)
        integral = 0.0
    
    for i in range(N):
        integral += func(xrand[i])


    answer = (b-a)/float(N)*integral

    areas.append(answer) 

plt.title("Distribution of Areas")
plt.hist(areas, bins = 30, ec = "black")
plt.xlabel("Areas")
plt.plot()
```

#### Q4. In python, plot the probability mass function and cumulative distribution function of the following random variables

(a) Binomial, n = 30, p = 0.2


```python
X = stats.binom(30, 0.2) # Declare X to be a binomial random variable
arr = [x for x in range(1, 31)]
answer = []
for i in range(1, 31):
    answer.append(X.cdf(i))  # P(X <= i)

plt.plot([x for x in range(1, 31)], answer)
plt.suptitle('Cumulative Distribution')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.show()
```

```python
answer = []
for i in range(1, 31):
    answer.append(X.pmf(i))  # P(X <= i)
    
val = [x for x in range(1, 31)]
plt.bar(val, answer)
plt.suptitle('Mass Distribution')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.show()
```
